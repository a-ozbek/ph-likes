```python
import os
import pickle
import numpy as np
import pandas as pd
from tqdm import tqdm, tqdm_notebook
from ph_likes.util import queries
from ph_likes.util import settings
```

```python
# df_user = queries.get_phuser_table()
df_post = queries.get_post_table()
df_postlike = queries.get_postlike_table()
```

```python
df_postlike
```

```python
# post_id index
unique_post_ids = df_postlike['post_id'].unique()
post_ids2index = dict(zip(unique_post_ids, range(len(unique_post_ids))))
index2post_ids = dict(zip(post_ids2index.values(), post_ids2index.keys()))
```

```python
# user_id index
unique_user_ids = df_postlike['user_id'].unique()
user_ids2index = dict(zip(unique_user_ids, range(len(unique_user_ids))))
index2user_ids = dict(zip(user_ids2index.values(), user_ids2index.keys()))
```

```python
print('len(unique_post_ids):', len(unique_post_ids))
print('len(unique_user_ids):', len(unique_user_ids))
```

```python
# create Collaborative Filtering Matrix 
CF = np.full((len(unique_user_ids), len(unique_post_ids)), False, dtype=np.bool)
CF.shape
```

```python
for i, row in tqdm_notebook(df_postlike.iterrows(), total=len(df_postlike)):
    i_user_id = user_ids2index[row['user_id']]
    i_post_id = post_ids2index[row['post_id']]
    CF[i_user_id, i_post_id] = True
```

# Save

```python
np.save(os.path.join(settings.LOCAL_MODELS_PATH, 'CF.npy'), CF)
```

```python
# post_ids
pickle.dump(post_ids2index, open(os.path.join(settings.LOCAL_MODELS_PATH, 'post_ids2index.pkl'), 'wb'))
pickle.dump(index2post_ids, open(os.path.join(settings.LOCAL_MODELS_PATH, 'index2post_ids.pkl'), 'wb'))
```

```python
# user_ids
pickle.dump(user_ids2index, open(os.path.join(settings.LOCAL_MODELS_PATH, 'user_ids2index.pkl'), 'wb'))
pickle.dump(index2user_ids, open(os.path.join(settings.LOCAL_MODELS_PATH, 'index2user_ids.pkl'), 'wb'))
```
